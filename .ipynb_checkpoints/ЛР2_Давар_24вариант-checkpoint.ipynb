# Подключение нужных для работы библиотек

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# __Чтение файла и вывод датафрейма__

```python
df = pd.read_csv('visits2.csv',sep = ';')
df
```

```python
# Информация о наборе данных visits2.csv.

Данные пользовательских сессии магазина:

1. уникальный идентификатор пользователя
2. страна пользователя
3. устройство пользователя
4. идентификатор рекламного источника, из которого пришел пользователь
5. дата и время начала сессии
6. дата и время окончания сессии
7. время сессии в минутах
8. кол-во кликов пользователя
9. количество товаров в корзине
10. стоимость покупок
11. возраст пользователя
```

# Проверка информации о типах данных

```python
df.info()
```

Многие типы данных в наборе не соответствуют заданным требованиям. Большинство числовых значений представлено в виде чисел с плавающей запятой, тогда как ожидается целочисленный формат. Для корректировки данных потребуется преобразование типов с помощью команды astype().

```python
df['time_session'] = df['time_session'].astype('int64')
df['click_count'] = df['click_count'].astype('int64')
df['buy_count'] = df['buy_count'].astype('int64')
df['price'] = df['price'].astype('int64')

df['session_start'] = pd.to_datetime(df['session_start'], format='%d.%m.%Y %H:%M')
df['session_end'] = pd.to_datetime(df['session_end'], format='%d.%m.%Y %H:%M')

df.info()
```

# Проверка полей с помощью df.columns

```python
df.columns
```

Выявлена неточность в наименовании одного из полей. Код исправление представлен ниже.

```python
df = df.rename(columns={'user_id ': 'user_id'})
df.columns
```

# Просмотр всех пропущенных значений

```python
print(df.isna().sum())
df = df.dropna(subset = ['region'])
df = df.dropna(subset = ['device'])
print(df.isna().sum())
```


В данном случае количество пропущенных значений незначительно, поэтому их можно удалить без вреда для дальнейшего анализа.

# Проверка уникальных данных в таблице

```python
for i in df.columns:
    print(i,df[i].unique(), sep = '\n')
    print()
```


Обнаружены некорректные данные в region и device. Нужно привести их к единому формату c помощью команды replace().

```python
df['region'] = df['region'].replace('USA', 'United States')
print(df['region'].unique())

df['device'] = df['device'].replace({'MAC': 'Mac', 'IPHONE': 'iPhone'})
print(df['device'].unique())
```

# Проверка явных дубликатов.

```python
print(df.duplicated().sum())
```

По результатам проведенной проверки явных дубликатов не было обнаружено.

# Матрица диаграммы рассеяния

```python
df.columns
```

Построим матрицу.

```python
sns.pairplot(df[['time_session','click_count','buy_count','price','age']])
```


На графике наблюдаются следующие взаимосвязи между переменными:

1. time_session и click_count: С увеличением времени сессии пользователь совершает больше кликов.

2. time_session и buy_count: Чем дольше пользователь находится в системе, тем больше покупок он совершает.

3. time_session и price: С увеличением времени сессии пользователи склонны к увеличению общей стоимости покупок.

4. click_count и buy_count: С ростом количества кликов пользователь совершает больше покупок.

5. click_count и price: Чем больше кликов совершает пользователь, тем выше общая стоимость его покупок.

6. buy_count и price: С увеличением количества покупок пользователем возрастает и общая стоимость покупок.


# Исследование взаимосвязей между переменными было проведено с использованием оценки коэффициентов корреляции и ковариации.

Для более глубокого анализа взаимосвязей между переменными вычисляются коэффициенты корреляции и ковариации. Корреляционная матрица вычисляется с помощью метода corr(), а ковариационная матрица — с помощью метода cov(). Для визуализации корреляционной матрицы строится тепловая карта с помощью heatmap() из библиотеки Seaborn. Это позволяет наглядно представить взаимосвязи между переменными.

Расчёт коэффициента корреляции и построение тепловой карты.

```python
correlation_matrix = df[['time_session','click_count','buy_count','price','age']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=1)
plt.title("Тепловая карта корреляции")
plt.show()
```

Выводы корреляции:

Результаты анализа показали, что все переменные, за исключением age, имеют высокие значения коэффициента корреляции, равные 0.87 и выше. Это свидетельствует о наличии сильной линейной взаимосвязи между ними: увеличение одной переменной сопровождается увеличением других.

Коэффициенты корреляции для переменной age близки к 0, что указывает на отсутствие значимой линейной взаимосвязи с остальными параметрами. Это означает, что изменения в возрасте пользователя не влияют на время сессии, количество кликов и покупок или стоимость.

Расчёт коэффициента ковариации.

```python
covariation_matrix = df[['time_session','click_count','buy_count','price','age']].cov()
print(covariation_matrix)
```

Выводы ковариации:

Все значения ковариации для time_session положительные, кроме ковариации с age, что свидетельствует о том, что увеличение времени сессии связано с увеличением количества кликов, покупок и общей стоимости покупок.

Для click_count также наблюдаются положительные значения ковариации, что указывает на то, что увеличение количества кликов связано с ростом времени сессии, количества покупок и ценой.

Значение ковариации между buy_count и price также положительное, что подтверждает, что увеличение количества покупок связано с увеличением общей стоимости покупок.

Все значения ковариации для age с другими переменными отрицательные. Это указывает на наличие обратной взаимосвязи: по мере увеличения возраста пользователя время сессии, количество кликов, покупок и общая стоимость имеют тенденцию к снижению.

Ковариация для age по сравнению с остальными переменными имеет относительно малые значения, что указывает на то, что изменения в возрасте пользователя не сильно влияют на другие параметры.

# Группировка данных по регионам и каналам с последующим подсчётом количества пользователей в каждой группе.

```python
group = df.groupby(['region', 'channel'])['user_id'].count().reset_index(name='count')
print(group)
```

# Построение графика.

```python
sns.barplot(data = group, x='channel', y ='count',hue='region')
plt.title('Диаграмма количества клиентов по каждому региону и каналу привлечения')
plt.xlabel('Канал привлечения')
plt.ylabel('Количество клиентов')
plt.legend(title='region')
plt.grid()
plt.show()
```

Согласно диаграмме, наблюдается доминирование United States по количеству клиентов во всех каналах привлечения. Наибольшее количество клиентов было привлечено через organic (более 500), за ним следуют FaceBoom и TipTop, а наименьшее количество клиентов пришло из MediaTornado.

Также преобладание клиентов из США подчеркивает необходимость пересмотра маркетинговых стратегий для привлечения пользователей из других стран, включая Россию. Анализ показателей между регионами может помочь выявить успешные методы и определить, какие каналы наиболее эффективны для каждого рынка.

# Создание сводной таблицы, отражающей количество уникальных пользователей для каждого канала.

```python
pivot = df.pivot_table(index='channel', values='user_id', aggfunc='nunique')
print(pivot)
```

# Построение графика.

```python
pivot.plot(kind = 'line', y = "user_id",style = 'blue', marker='<', linestyle='None', markersize=15,  grid = 2, title = "График уникального количества пользователей по каналам привлечения", ylabel ='Количество пользователей',xlabel = "Канал привлечения", figsize = (8,6))
plt.show()
```

Наибольшее количество уникальных пользователей было привлечено через organic, затем идут FaceBoom и TipTop, а на последнем месте находится MediaTornado.

Эта информация указывает на высокую эффективность organic в привлечении пользователей, что подчеркивает значимость качественного контента и SEO-оптимизации. Также видно, что FaceBoom и TipTop показывают хорошие результаты, что свидетельствует о потенциале для их дальнейшего роста. В то же время MediaTornado требует пересмотра стратегии для повышения привлекательности и эффективности.

# Группировка device по размеру.

```python
group_device = df.groupby(['device']).size()
print(group_device)
```

# Построение графика.

```python
plt.figure(figsize=(20, 10))
plt.pie(group_device, labels = group_device.keys(), autopct='%1.1f%%')
plt.legend(fontsize = 20, bbox_to_anchor=(1, 1))
plt.title('Распределение устройств')
plt.show()
```


На графике представлено распределение устройств: наибольшую долю занимает iPhone (44,1%), за ним следует Mac (25,6%), затем Android (19,4%), а на последнем месте располагается PC (10,8%).

Из данной информации можно сделать вывод о значительном преобладании мобильных устройств, особенно iPhone, в использовании по сравнению с другими платформами. Это может свидетельствовать о предпочтении пользователей к мобильным устройствам для доступа к контенту или услугам.
